## Problem 2: Arithmetic

The **Rule of 70** says that at an annual growth rate of $g$ percent, a
quantity doubles in approximately $70/g$ years.

**(a)** Using Julia, compute how many years it takes GDP to double at
growth rates of 2%, 3%, and 7%.

```julia
for g in [2, 3, 7]
    println("At $g% growth, the Rule of 70 says doubling takes $(70/g) years.")
end
```

**(b)** Check the Rule of 70 against the exact formula. The exact
doubling time is $\ln(2)/\ln(1 + g/100)$. Compute the exact doubling
time for the same three growth rates. How good is the approximation?

*Hint:* The natural log function in Julia is `log()`.

```julia
for g in [2, 3, 7]
    rule70 = 70 / g
    exact = log(2) / log(1 + g/100)
    println("g = $g%: Rule of 70 = $(round(rule70, digits=2)), Exact = $(round(exact, digits=2))")
end
```

The Rule of 70 is a very good approximation for low growth rates (2-3%)
and slightly less accurate but still reasonable at higher rates (7%).

## Problem 3: Arrays

**(a)** Create an array `prices` containing the values
`[4.50, 3.25, 7.00, 2.75, 5.50]` (prices of five goods). Compute the
mean price using a `for` loop (do not use the built-in `mean` function).

```julia
prices = [4.50, 3.25, 7.00, 2.75, 5.50]

total = 0.0
for p in prices
    total += p
end
mean_price = total / length(prices)
println("Mean price: $mean_price")
```

**(b)** Using array slicing, extract the first three prices into a new
array called `cheap`. Then use `push!` to add the value `1.99` to the
end of `cheap`.

```julia
cheap = prices[1:3]
push!(cheap, 1.99)
println(cheap)
```

**(b)** Create a dictionary called `gdp_data` that maps country names
(strings) to GDP values: `"US" => 25.46`, `"China" => 17.96`,

```julia
gdp_data = Dict("US" => 25.46, "China" => 17.96, "Japan" => 4.23, "Germany" => 4.07)
gdp_data["UK"] = 3.07
println(gdp_data)
```

**(c)** Write a `for` loop that iterates over `gdp_data` and prints each
